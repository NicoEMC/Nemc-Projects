## Se nos pide graficar la cantidad de contagios por año según mes correspondiente 
**El data set contiene 5 años de registros desde el 2017 al 2021**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%config IPCompleter.greedy = True
```

```python
df = pd.read_csv('Contagiados_1.csv')
```

```python
df.head(13)
```

```python
### Separamos el mes y el dia, luego creamos dos columnas nuevas
df_mes = df['Mes'].str.split('-',expand = True)
df_mes.columns = ['Dia_mes','Mes_año']

df_mes.head(5)
```

```python
### Unimos las nuevas dos columnas al dataframe original
df = pd.concat([df,df_mes], axis =1)
df.head() 
```

```python
### Generamos una lista con los meses del Año, tambien se puede realizar con mes = list(df['Mes_año'])
#mes_prueba = list(df['Mes_año'])
#mes_prueba
Mes = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
```

```python
### Transformamos el mes en variable categorica (Mes_año) y lo ordenamos según la lista entregada
df['Mes_año'] = pd.Categorical(df['Mes_año'], categories = Mes, ordered = True )
df = df.sort_values('Mes_año')
```

```python
df.head(12)
```

```python
df.tail(12)
```

```python
### Creamos un Plot de todos los meses con cada año agrupado
plt.figure(figsize = (20,10))

plt.suptitle('Grafico', fontsize = 14, color='black', y= 1.05)
grid_bar_n = sns.FacetGrid(df , col = 'Mes', col_wrap = 4 , hue = 'Mes')

### map nos permite iterar cada uno de los meses según año y fallecidos
grid_bar_n.map(plt.barh, 'Año', 'Fallecidos')


plt.show()
```
