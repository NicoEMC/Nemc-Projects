```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Caso 1

* ¿Cuál paleta es más efectiva para codificar estos datos? R: La paleta divergente porque es más fácil detectar outliers, como el lunes de la última semana o el domingo de la primera, o se puede saber fácilmente si se esta alejando del 'centro' de los datos

```python
df_video = pd.read_csv('./videoconferencia.csv').pivot('Día', 'Semana', 'Horas').reindex(['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'])

fig1, (ax1, ax2) = plt.subplots(2, 1, figsize=(20,30))
sns.heatmap(data=df_video, ax=ax1, cmap='vlag', annot=True, fmt='0')
ax1.set_title('CMap divergente (vlag)')
sns.heatmap(data=df_video, ax=ax2, cmap='mako', annot=True, fmt='0')
ax2.set_title('CMap secuencial (mako)')
```

## Caso 2

* a) de los supervivientes, ¿cuál es la clase de pasajero más común? R: Pasajeros de primera clase
* b) de los no supervivientes ¿cuál es la clase de pasajero más común? R: Pasajeros de tercera clase
* c) ¿cree que este gráfico es más efectivo que un gráfico de barras para comparar categorías? R: No, según Munzner, para estos casos, un canal de una dimensión (una línea o, según la pregunta, la longitud de una barra) es más efectivo que un área (canal de dos dimensiones)

```python
df_titanic = sns.load_dataset('titanic')

# TODO: Buscar una mejor forma de hacer esto
df_class = df_titanic.groupby('pclass').agg({'class': len})
df_class.index = ['First', 'Second', 'Third']
df_survived = df_titanic.loc[df_titanic['survived'] == 1]
df_died = df_titanic.loc[df_titanic['survived'] == 0]
df_survived = df_survived.groupby('pclass').agg({'class': len})
df_died = df_died.groupby('pclass').agg({'class': len})
df_survived.index = ['First', 'Second', 'Third']
df_died.index = ['First', 'Second', 'Third']

fig2, axes = plt.subplot_mosaic([['A', 'A'], ['B', 'C']], figsize=(10, 10))
fig2.tight_layout()
df_class.plot(kind='pie', y='class', ax=axes['A'], autopct='%1.1f%%', startangle=90)
df_died.plot(kind='pie', y='class', ax=axes['B'], autopct='%1.1f%%', startangle=180)
df_survived.plot(kind='pie', y='class', ax=axes['C'], autopct='%1.1f%%', startangle=180)

for _, ax in axes.items():
    ax.get_legend().remove()
    ax.set_ylabel('')
    
axes['A'].set_title('Pasajeros del Titanic por clase')
axes['B'].set_title('Pasajeros fallecidos del Titanic por clase')
axes['C'].set_title('Pasajeros sobrevivientes del Titanic por clase')
```

## Caso 3

* a) ¿cuál de los dos gráficos permite comparar la tendencia de un género entre los diferentes años?  R: El gráfico de líneas, puesto que cada gráfico hace comparaciones sólo dentro de un mismo género considerando los años.
* b) ¿cual de los dos gráficos permite identificar el género de películas más estrenados en 1 año específico? R: El gráfico de barras horizontales, ya que este compara los estrenos de cada género en cada año

```python
grid1 = sns.FacetGrid(data=df_movies, col='género', col_wrap=3, hue='género')
grid1.map(plt.plot, 'año', 'cantidad', marker='*')
```
