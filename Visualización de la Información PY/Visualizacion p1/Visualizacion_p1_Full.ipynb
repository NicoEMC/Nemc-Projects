## Caso 1: Videoconferencia

```python
df = pd.read_csv('videoconferencia.csv' )
df.tail()
```

```python
visual = df.pivot(index= 'Día', columns ='Semana', values ='Horas')
```

```python
Dias = ['Lunes','Martes','Miércoles','Jueves','Viernes','Sábado','Domingo']
```

```python
visual = visual.reindex(Dias)
visual
```

```python
ax = sns.heatmap(visual)
```

```python

plt.figure(figsize =(20,9))

## Tamaño letra
sns.set(font_scale = 1)

ax = sns.heatmap(visual, annot=True)
```

```python
plt.figure(figsize =(20,9))
sns.set(font_scale = 1)

ax = sns.heatmap(visual, annot=True, fmt='d', linewidths=1, square=True)
```

```python
### PALETA DE COLORES SECUENCIAL 
plt.figure(figsize =(20,9))
sns.set(font_scale = 1)

ax1 = sns.heatmap(visual, annot=True, fmt='d', linewidths=0.8, cmap='YlOrBr')
ax1.set_title('Color Paleta Secuencial "YlOrBr" ')
```

```python
### PALETA DE COLORES SECUENCIAL INVERTIDA
plt.figure(figsize =(20,9))
sns.set(font_scale = 1)

ax1 = sns.heatmap(visual, annot=True, fmt='d', linewidths=0.8, cmap='YlOrBr_r')
ax1.set_title('Color Paleta Secuencial "YlOrBr_r" ')
```

```python
### PALETA DE COLORES DIVERGENTE 
plt.figure(figsize =(20,9))
sns.set(font_scale = 1)

ax2 = sns.heatmap(visual, annot=True, fmt='d', linewidths=0.8, cmap="coolwarm")
ax2.set_title('Color Paleta Secuencial "coolwarm" ')
```

```python
### PALETA DE COLORES DIVERGENTE INVERTIDA
plt.figure(figsize =(20,9))
sns.set(font_scale = 1)

ax2 = sns.heatmap(visual, annot=True, fmt='d', linewidths=0.8, cmap="coolwarm_r")
ax2.set_title('Color Paleta Secuencial "coolwarm_r" ')
```

## Caso 2: Titanic

```python
df_titanic = sns.load_dataset("titanic")
```

```python
df_titanic.tail()
```

```python
## Columnas del dataset
df_titanic.columns
```

```python
## Agrupar por clase de pasajero
df_tit_total = df_titanic.groupby('class').count()
df_tit_total
```

```python
## Se crea la figura
fig = plt.figure(figsize=(20,9))

## creamos el entorno
plot = fig.add_subplot()

clases = ["First","Second","Third"]

colores = sns.color_palette("Set2")

## grafico de torta
pie_plot = plot.pie(df_tit_total["pclass"], labels = clases, autopct="%0.1f %%",startangle = 90, colors = colores)

## Leyenda
# plot.legend(pie_plot[0], df_tit_total['survived'], title="Sobrevivientes por clase")

## Titulo
plot.set_title("Pasajeros del Titanic por clase")

plt.axis("equal")

plt.show()
```

```python
## Identificamos los pasajeros No sobrevivientes 
df_titanic.loc[df_titanic['survived']==0]
```

```python
## Contamos la cantidad de cada clase
df_titanic.loc[df_titanic['survived']==0].groupby(['class']).count()
```

```python
## Asignamos la variable a cada situación (Sobrevivientes y muertos)
Muertos = df_titanic.loc[df_titanic['survived']==0].groupby(['class']).count()
Sobrevivientes = df_titanic.loc[df_titanic['survived']==1].groupby(['class']).count() 
```

```python
Muertos['survived']
```

```python
## Graficamos por Pasajeros del Titanic si sobrevivio o no el pasajero y por clase de pasajero
fig2, (ax1, ax2) = plt.subplots(1,2, figsize=(15,9))
## ajuste para no empalme de gráficos
fig2.tight_layout()
fig2.suptitle('Pasajeros del Titanic por Superviviencia (0/1) y por Clase (T/F/S)')
#ax.pie(df_tit_total)
ax1.pie(Muertos['survived'], autopct="%0.1f %%",startangle = 180, colors = ['blue','orange','green'], labels= clases)
ax1.set_ylabel('0',fontweight='bold')
ax1.set_title('Pasajeros fallecidos del Titanic por clase')

ax2.pie(Sobrevivientes['survived'], autopct="%0.1f %%",startangle = 180, colors = ['blue','orange','green'], labels= clases)
ax2.set_ylabel('1',fontweight='bold')
ax2.set_title('Pasajeros sobrevivientes del Titanic por clase')


```

#### Caso 2: Respuesta 

- a) De los supervivientes ¿Cuál es la clase de pasajero más común? 

La clase que es más común de las personas supervivientes en el dataset de "Titanic.csv" son los Pasajeros de primera clase, con un porcentaje del 39.8% del total de clase de personas.


- b) De los no supervivientes ¿Cuál es la clase de pasajero más común?

La clase que es más común de las personas NO supervivientes en el dataset de "Titanic.csv" son los Pasajeros de tercera clase , con un porcentaje del 67.8% del total de clase de personas.


- c) ¿Cree que este gráfico es más efectivo que un gráfico de barras para comparar categorías?

No debería ser más efectivo, dado que si no se utilizarán las leyendas no se podría apreciar la diferencia de volumen en cada pedazo de las tortas, en el caso de un gráfico de barrás podríamos notar a simple vista según la longitud de la barra quién es el que tiene mayor o menor valor en cada caso.



```python
## Gráficos agrupados 
### Figura para 3 gráficos 
fig3, axes = plt.subplot_mosaic([['Clase','Clase'],['Muertos','Sobrevivientes']], constrained_layout = True,figsize=(12,12))

### Creamos el primer gráfico, le colocamos título y le eliminamos la etiqueta en Y
df_tit_total.plot(kind='pie', y='pclass',ax = axes['Clase'], legend=None, autopct="%0.1f %%",
                  startangle = 90, colors = ['c','orange','green'], fontsize = 18)

axes['Clase'].set_title('Pasajeros del Titanic por clase', fontsize = 25)
axes['Clase'].set_ylabel(None)

### Creamos un Título y lo posicionamos en el centro de los gráficos
fig3.suptitle('Pasajeros del Titanic por Supervivencia (0/1) y por Clase (T/F/S)', x = 0.5, y = 0.5,
              verticalalignment = 'center', fontsize = 20)

### Creamos el segundo gráfico, le colocamos título, le añadimos la enumeración de los pasajeros 
###     muertos como '0' en la etiqueta en Y y eliminamos el título
Muertos.plot(kind='pie', y='pclass',ax = axes['Muertos'], legend=None, autopct="%0.1f %%",
             startangle = 180, colors = ['c','orange','green'], labels= clases, fontsize = 18)

axes['Muertos'].set_title(None)
axes['Muertos'].set_ylabel('0',fontweight='bold', fontsize = 20)

### Creamos el tercer gráfico, le colocamos título, le añadimos la enumeración de los pasajeros 
###     sobrevivientes como '1' en la etiqueta en Y y eliminamos el título
Sobrevivientes.plot(kind='pie', y='pclass',ax = axes['Sobrevivientes'], legend=None, autopct="%0.1f %%",
                    startangle = 180, colors = ['c','orange','green'], labels= clases, fontsize = 18)

axes['Sobrevivientes'].set_title(None)
axes['Sobrevivientes'].set_ylabel('1',fontweight='bold', fontsize = 20)

plt.show()
```

## Caso 3: Peliculas

```python
df_peliculas = pd.read_csv('peliculas.csv')
df_peliculas
```

```python
len(df_peliculas['género'])
```

```python
## Gráfico de líneas
grid_linea = sns.FacetGrid(df_peliculas , col = 'género', col_wrap = 3 , hue = 'género')
grid_linea.map(plt.plot, 'año', 'cantidad', marker = 'o')
```

```python
## Gráfico de Barras horizontal
grid_bar = sns.FacetGrid(df_peliculas , col = 'año', col_wrap = 3 , hue = 'año')
grid_bar.map(plt.barh, 'género', 'cantidad')
```

#### Caso 3: Respuesta

 - ¿Cuál de los dos gráficos permite comparar la tendencia de un género entre los diferentes años? 
 
 El gráfico de líneas nos permite comparar la tendencia, ya que este gráfico considerá el valor de la cantidad de películas por género que han sido estrenadas a tráves de los años, y con esto, podemos hacer comparaciones visualizando un gráfico a la vez por genero.
 
 
 - ¿Cuál de los dos gráficos permite identificar el género de películas más estrenados en 1 año específico?
 
 El gráfico de Barras horizontales nos permite visualizar que género de peliculas fue el que más estrenos tuvo en comparación a los demás generos en cada año.

```python
## Gráficos Agrupados
grid_linea = sns.FacetGrid(df_peliculas , col = 'género', col_wrap = 3 , hue = 'género')
grid_linea.map(plt.plot, 'año', 'cantidad', marker = 'o')

grid_bar = sns.FacetGrid(df_peliculas , col = 'año', col_wrap = 3 , hue = 'año')
grid_bar.map(plt.barh, 'género', 'cantidad')
```
