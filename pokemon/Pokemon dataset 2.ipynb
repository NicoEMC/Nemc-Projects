
### 


* Se nos pide crear una variable dentro del dataset que contenga el nombre y la vida (HP) en una misma variable o columna



```python
"""
zip agrupa dos o más variables en una tupla, por lo que primero pasamos toda la columna del dataset a una lista
luego realizamos el join entre listas con zip, el '[*]' nos permite crear una lista con todos los valores.



"""
df['Nombres_hps'] = [*(zip(list(df['Name']),list(df['HP'])))]
```

```python
df['Nombres_hps']
```

```python
df.tail()
```

```python
### Podemos visualizar la combinación de valores de la columna recien creada
for x,y in df['Nombres_hps'][:10]:
    print("Pokemon: {}, Hp: {}".format(x,y)),
    print("Hp: {}".format(y)),
    print("Hp: {},Pokemon: {}".format(y,x))
```

### Contar cada uno de los tipos de Pokemon según el Type 1 class

```python
df.head()
```

```python
from collections import Counter
poke_tipo = df['Type 1']
```

```python
tipos_cuenta = Counter(poke_tipo)
print(tipos_cuenta)
```

```python
### Contar con Loop, creando un diccionario
"""
Por cada tipo se crea uno de cada uno, luego al ya existir, se adiciona un 1 a la cuenta anterior

"""

tipos_cuenta_2 = {}

for tipo in poke_tipo:
    if tipo not in tipos_cuenta_2:
        tipos_cuenta_2[tipo] = 1
    else:
        tipos_cuenta_2[tipo] +=1
print(tipos_cuenta_2)
```

### Contar numero de letras de cada nombre Pokemon

- Reduciremos la muestra a solo pokemones de la generación 1, 2 y 3 

```python
df.info()
```

```python
df_gen = df[df['Generation'] < 4]
df_nombres = df_gen['Name']
#df_gen_lista = list(df_gen)
```

```python
df_nombres
```

```python
largo_nombre = map(len, df_nombres)
```

```python
longitud_nombre = [*zip(df_nombres, largo_nombre)]
```

```python
print(longitud_nombre[:5])
```

```python
### TOP 3 Con menor y mayor cantidad de letras

top_3_min = sorted(longitud_nombre, key=lambda x: x[1], reverse=False)[:3]
top_3_max = sorted(longitud_nombre, key=lambda x: x[1], reverse=True)[:3]
print("Los pokemones con menos letras en el nombre son: ",top_3_min)
print("Los pokemones con más letras en el nombre son: ",top_3_max)
```
