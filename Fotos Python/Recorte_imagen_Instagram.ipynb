```python
### Seleccionamos el nombre de la imagen
img = "insta.jpeg"
```

```python
### Abrimos la imagen y vemos las dimensiones de ella
img = Image.open(img)
img.size
```

```python
## Imagen original
plt.figure(figsize = (16,12))
imgplot = plt.imshow(img)
plt.show()
```

```python
### Utilizamos crop para recortar la imagen, debemos ir probando para llegar a la imagen deseada
### crop(izquierda,arriba,derecha,abajo) <-- sectores que queremos cortar en pixeles
region = (0,140,540,740)
img_recortada = img.crop(region)

#img_recortada.show()
plt.figure(figsize = (16,12))
imgplot = plt.imshow(img_recortada)
plt.show()
print(img.size)
```
